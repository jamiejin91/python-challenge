```python
import re
import time
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import requests
```

```python
# Get all links of articles for all pages
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser("chrome", **executable_path, headless=True)

browser.visit('http://sn4hr.org/?s=school')
time.sleep(1)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
last_page = int(soup.find('span', class_='pages').text.split(' ')[-1])

count = 1
links_all = []
for page in range(1,last_page+1):

    html_page = requests.get('http://sn4hr.org/page/{}/?s=school'.format(page))
    soup_page = BeautifulSoup(html_page.text, 'html.parser')
    links = soup_page.find_all('h3', class_='entry-title td-module-title')

    for link in links:
        try: 
            links_all.append(link.a['href'])
            print("Article {} added".format(count))
        except:
            print("Article {} failed".format(count))
        count+=1
    
    print("page {} complete".format(page))
```

```python
# Parsing each article
json_final = []
for link in links_all:
    html_page = requests.get(link)
    soup_page = BeautifulSoup(html_page.text, 'html.parser')
    
    # title of article
    title = soup_page.find('h1', class_="entry-title").text.strip()
    
    # content
    content_temp = soup_page.find('h5').text.strip()
    pattern = re.compile('([^\s\w]|_)+')
    content = pattern.sub('', content_temp)
    
    # aggressor
    aggro = title.split('shelled')[0]
    
    # date
    date = content.split(' ')[-3:]
    month_num = int(time.strptime(date[0],'%B').tm_mon)
    
    # append to json list
    json_final.append({
        "date":"{}-{}-{}".format(int(date[2]),mont_num,int(date[1])),
        "title":title,
        "content":content,
        "aggressor":aggro
                      })
    
    ## govoronate: between "in" and date[0] minus "on" or ",", or match from syria_locale.py
    ## municipality: between "in" and "in"
    ## school: between "targeted"/"shelling"/"shelled" and "School+
    ## probably best making a list of places/locations/forces etc.
    ## ML...?
```
