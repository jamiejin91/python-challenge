```python
import pandas as pd
```

```python
df_m = pd.read_csv('Resources/hawaii_measurements.csv')
df_m.fillna(0, inplace=True)
df_m['date'] = df_m['date'].apply(lambda x: x[:4]+x[5:7]+x[8:])
df_m['date'] = df_m['date'].astype(int)
df_m.to_csv('Resources/clean_hawaii_measurements.csv', index=False)
df_m
```
